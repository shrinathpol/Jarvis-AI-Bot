# Data Exploration

This notebook is used for exploratory data analysis (EDA) on the training and validation datasets. It includes visualizations and insights derived from the data.

```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns

# Load training data
with open('../data/training_data.json') as f:
    training_data = json.load(f)

# Load validation data
with open('../data/validation_data.json') as f:
    validation_data = json.load(f)

# Convert to DataFrame for analysis
training_df = pd.DataFrame(training_data)
validation_df = pd.DataFrame(validation_data)

# Display the first few rows of the training data
training_df.head()
```

```python
# Summary statistics of the training data
training_df.describe()
```

```python
# Visualize the distribution of a specific feature
plt.figure(figsize=(10, 6))
sns.histplot(training_df['training_data'].apply(lambda x: x['feature_1']), bins=30, kde=True)
plt.title('Distribution of Feature 1')
plt.xlabel('Feature 1')
plt.ylabel('Frequency')
plt.show()
```

```python
# Extract features into separate columns
training_df['feature_1'] = training_df['training_data'].apply(lambda x: x['feature_1'])
training_df['feature_2'] = training_df['training_data'].apply(lambda x: x['feature_2'])
training_df['label'] = training_df['training_data'].apply(lambda x: x['label'])

# Calculate the correlation matrix
correlation_matrix = training_df[['feature_1', 'feature_2', 'label']].corr()

# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```
